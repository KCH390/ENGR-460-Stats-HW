```python
#sleep duration data dictionary gathered in class
sleep_dict = {
    "Bouba" : 5,
    "Harrison" : 5,
    "Andrew" : 5,
    "Chase" : 5, 
    "Ivey" : 6.5,
    "Garrett" : 6,
    "Enrique" : 6.5,
    "Josh H." : 5.5,
    "Chris" : 5,
    "Greeley" : 6.5,
    "Sam" : 8.5,
    "Nick" : 10,
    "Zach S." : 7,
    "Zach L." : 6,
    "Pace" : 4,
    "Mattie" : 7,
    "Stephen" : 9,
    "Jack" : 8,
    "Josh" : 8,
    "Matthew" : 5.5,
    "Kerry" : 8.25,
    "Chris Z" : 5,
    "Alejandro" : 7.5,
    "Josh M" : 8,
    "Emmanuel" : 6,
}
```

```python
#building the initial discrete lists
set_less3 = [] 
set_less6 = []
set_less9 = []
set_great9 = []
for each in sleep_dict:
    if sleep_dict.get(each) < 3: set_less3.append(sleep_dict.get(each)) # condition for less3 list
    elif sleep_dict.get(each) < 6: set_less6.append(sleep_dict.get(each)) # condition for less6 list
    elif sleep_dict.get(each) < 9: set_less9.append(sleep_dict.get(each)) # condition for less9 list
    else: set_great9.append(sleep_dict.get(each)) # everything else must be greater than 9 so they go here

print("The sample spaces are:", set_less3, set_less6, set_less9, set_great9, sep="\n\n") #printing output of for loop
mu = (sum(set_less3)+sum(set_less6)+sum(set_less9)+sum(set_great9))/len(sleep_dict) # calculating the average amount of sleep as a continuous function
variance = std(list(sleep_dict.values()),ddof=1)**2 #ASSUMING SAMPLE variance AND NOT POPULATION variance
print("\nThe average amount of sleep time is: ", mu, " hours.", " The sample variance is: ", round(variance, 3), " hours^2.", sep="") #printing average and variance
mag_vec = array([len(set_less3), len(set_less6), len(set_less9), len(set_great9)]) # doing this because it simplifies plotting
```

Based on the data gathered, the sleep duration distribution appears to be a non-normal distribution. This is because the vast majority of the values fall between 6 and 9. The sample size is too small to say that this is certain. The experiment could be improved by gathering data from additional people over a longer period of time. 

```python
fig, ax = plt.subplots(figsize=(17,13)) # initialization
ax.bar(range(4),mag_vec/sum(mag_vec), edgecolor='black',color=["#FF0000","#FF0000","#00FF00","#00FF00",]) #construction of a bar plot
plt.xticks(range(0,len(["0 to 3", "3 to 6", "6 to 9", "9 or more"])), ["0 to 3", "3 to 6", "6 to 9", "9 or more"],fontsize=28) # using text labels for each column over default numeric index
ax.set_ylabel("Frequency Fraction",fontsize=24) # Labeling
ax.set_xlabel("Hours Spent Sleeping (endpoints not inclusive)",fontsize=24) # Labeling
plt.title("Relative Sleep Duration Frequency",fontsize=24) # Labeling
plt.show() # printing
```

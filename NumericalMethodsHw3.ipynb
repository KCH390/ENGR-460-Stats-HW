```python
#Unless otherwise noted, code present was constructed by and is the explicit property of Kerry Hall. All conclusions present are property of Kerry Hall. 
import numpy as np
from numpy import array, zeros, fabs
import time
import scipy.interpolate as sci
import matplotlib.pyplot as plt
```

```python
data_x = np.array([n for n in range(4)])
data_y = np.array([3, 5, 4, 1,])
test_x = np.array([0.5, 1.5])
xrange = np.linspace(0, 3., 70)
Interpolation_funcs = [sci.interp1d(data_x, data_y), sci.lagrange(data_x, data_y), sci.Akima1DInterpolator(data_x, data_y), sci.CubicSpline(data_x, data_y)]
m01 = (data_y[0] - data_y[1])//(data_x[0] - data_x[1]) # All integer values, couldn't use floordiv here if I didn't already know that these were integers.
m12 = (data_y[1] - data_y[2])//(data_x[1] - data_x[2])
m23 = (data_y[2] - data_y[3])//(data_x[2] - data_x[3])
print(f"\nThe interpolation equations are: \ny01 =  {m01}[x01] + {data_y[0]}; \ny12 = {m12}[x12] + {data_y[1] - m12*data_x[1]}; \ny23 = {m23}[x23] + {data_y[2] - m23*data_x[2]}. ")
```

For each of the interpolation functions, the solutions were placed before the plot. This was to ensure that the solutions were visible with the associated code.

```python
tic = time.perf_counter()
print("\nThe linear interpolation to x = 0.5 and x = 1.5 is ", Interpolation_funcs[0](test_x[0]), " and ", Interpolation_funcs[0](test_x[1]), " respectively.\n", sep="")
fig, ax = plt.subplots(figsize=(16,12))
ax.plot(data_x, data_y, color="#62f4c8", marker='h', linestyle='dashed', linewidth=3, markersize=15)
ax.scatter(test_x, Interpolation_funcs[0](test_x), color='#cf6fe9', s=135)
plt.title("Linear Interpolation of Dataset", fontsize=22)
ax.set_ylabel("y-data", fontsize=20)
ax.set_xlabel("x-data", fontsize=20)
plt.show()
toc = time.perf_counter()
lin_interp_t = toc - tic
```

```python
tic = time.perf_counter()
print("\n\nThe Lagrange interpolations to x = 0.5 and x = 1.5 are ", np.round(Interpolation_funcs[1](test_x[0]), 2), " and ", np.round(Interpolation_funcs[1](test_x[1]), 2), " respectively.\n", sep="")
fig, ax = plt.subplots(figsize=(16,12))
ax.plot(xrange, Interpolation_funcs[1](xrange), color="#2192FF", linestyle='dashed', linewidth=3)
ax.scatter(test_x, Interpolation_funcs[1](test_x), color='#38E54D', s=135)
ax.scatter(data_x, data_y, color="#2192FF",marker='h', s=135)
plt.title("Lagrange Interpolation of Dataset", fontsize=22)
ax.set_ylabel("y-data", fontsize=20)
ax.set_xlabel("x-data", fontsize=20)
plt.show()
toc = time.perf_counter()
lag_interp_t = toc - tic
```

```python
tic = time.perf_counter()
print("\n\nThe Akima Spline interpolations to x = 0.5 and x = 1.5 are ", np.round(Interpolation_funcs[2](test_x[0]), 2), " and ", np.round(Interpolation_funcs[2](test_x[1]), 2), " respectively.\n", sep="")
fig, ax = plt.subplots(figsize=(16,12))
ax.plot(xrange, Interpolation_funcs[2](xrange), color="#EB1D36", linestyle='dashed', linewidth=3)
ax.scatter(test_x, Interpolation_funcs[2](test_x), color='#73777B', s=135)
ax.scatter(data_x, data_y, color="#EB1D36",marker='h', s=135)
plt.title("Akima Spline Interpolation of Dataset", fontsize=22)
ax.set_ylabel("y-data", fontsize=20)
ax.set_xlabel("x-data", fontsize=20)
plt.show()
toc = time.perf_counter()
aki_interp_t = toc - tic
```

```python
tic = time.perf_counter()
print("\n\nThe Cubic Spline interpolations of x = 0.5 and x = 1.5 are ", np.round(Interpolation_funcs[3](test_x[0]), 2), " and ", np.round(Interpolation_funcs[3](test_x[1]), 2), " respectively.\n", sep="")
fig, ax = plt.subplots(figsize=(16,12))
ax.plot(xrange, Interpolation_funcs[3](xrange), color="#171717", linestyle='dashed', linewidth=3)
ax.scatter(test_x, Interpolation_funcs[3](test_x), color='#51C2D5', s=135)
ax.scatter(data_x, data_y, color="#171717",marker='h', s=135)
plt.title("Cubic Spline Interpolation of Dataset", fontsize=22)
ax.set_ylabel("y-data", fontsize=20)
ax.set_xlabel("x-data", fontsize=20)
plt.show()
toc = time.perf_counter()
cub_interp_t = toc - tic
```

```python
print(f"""\nThe linear interpolation took {np.round(1000*lin_interp_t, 2)} milliseconds.
          \n\nThe Lagrange interpolation took {np.round(1000*lag_interp_t, 2)} milliseconds.
          \n\nThe Akima spline interpolation took {np.round(1000*aki_interp_t, 2)} milliseconds.
          \n\nThe cubic spline interpolation took {np.round(1000*cub_interp_t, 2)} milliseconds.\n""")
```

What is the computing cost of each method?

The method that had the greatest time computing cost was the linear interpolation. The other three interpolations tended to have nearly the same computing cost, with variances in place. All three consistently beat linear interpolation, even with major computing blocks removed from the linear method code. 

What is the continuous situation at the node points of each method? 

The linear interpolation is continuous but only smooth in discrete intervals and not at nodes. The lagrange polynomial is continuous and nth order smooth everywhere. The Akima Spline is continuous and 1st order smooth at nodes. The cubic spline is continuous and 2nd order smooth at nodes.

What is the rounding error situation of each method? 

The Akima Spline has significant rounding error. The other three do not have significant rounding error relative to the Akima Spline.
